```python

import pandas as pd
import matplotlib.pyplot as plt


# Obtener url de archivo csv
path ="/content/atenciones_231005.csv"

# Leer el CSV
df = pd.read_csv(path)

# Obtener el conteo de cada categoría
conteo = df["anio_mes"].value_counts()


# Crear un gráfico de línea con dos líneas
plt.plot(conteo, conteo.index, color="blue")

plt.show()
```
